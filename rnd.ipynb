```python
y_pred_prob =  gbm.predict_proba(x_all)[:,1]
df['y_pred_P']=pd.DataFrame(y_pred_prob)
```

```python
df['P_rank_dtree']=pd.qcut(df['y_pred_P'].rank(method='first').values,10,duplicates='drop').codes+1
rank_df_actual= df.groupby('P_rank_dtree')['is_canceled'].agg(['count','mean'])
rank_df_predicted=df.groupby('P_rank_dtree')['y_pred_P'].agg(['mean'])
rank_df_actual=pd.DataFrame(rank_df_actual)
```

```python
rank_df_actual.rename(columns={'mean':'Actual_event_rate'},inplace=True)
```

```python
rank_df_predicted = pd.DataFrame(rank_df_predicted)
```

```python
rank_df_predicted.rename(columns={'mean':'Predicted_event_rate'},inplace=True)
rank_df = pd.concat([rank_df_actual,rank_df_predicted],axis=1, join="inner")
```

```python
sorted_rank_df = rank_df.sort_values(by='P_rank_dtree',ascending=False)
```

```python
sorted_rank_df['N_events']= rank_df['count']*rank_df['Actual_event_rate']
```

```python
sorted_rank_df['cum_events'] = sorted_rank_df['N_events'].cumsum()
```

```python
sorted_rank_df['events_cap'] = sorted_rank_df['N_events']/max(sorted_rank_df['N_events'].cumsum())
```

```python
sorted_rank_df['cum_event_cap'] = sorted_rank_df['events_cap'].cumsum()
```

```python
sorted_rank_df.shape
```

```python
sorted_rank_df['N_non_events'] = sorted_rank_df['count']-sorted_rank_df['N_events']
```

```python
sorted_rank_df['cum_non_events']= sorted_rank_df['N_non_events'].cumsum()
```

```python
sorted_rank_df['non_events_cap']=sorted_rank_df['N_non_events']/max(sorted_rank_df['N_non_events'].cumsum())
sorted_rank_df['cum_non_event_cap'] = sorted_rank_df['non_events_cap'].cumsum()
```

```python
sorted_rank_df['KS']= round((sorted_rank_df['cum_event_cap']-sorted_rank_df['cum_non_event_cap']),4)
```

```python
sorted_rank_df['random_cap']= sorted_rank_df['count']/max(sorted_rank_df['count'].cumsum())
```

```python
sorted_rank_df['cum_random_cap'] = sorted_rank_df['random_cap'].cumsum()
```

```python
sorted_reindexed= sorted_rank_df.reset_index()
```

```python
sorted_reindexed['Decile']= sorted_reindexed.index+1
```

```python
sorted_reindexed
```

```python
fig, axes = plt.subplots(1,2, sharex=True, figsize=(14,7))
fig.suptitle('Effectiveness Of Decile based on model Probabilities')
axes[0].set_title('Rank Ordering of actual event rate')
axes[1].set_title('Gains Chart ')

sns.lineplot(ax=axes[0], x="Decile", y="Actual_event_rate", data= sorted_reindexed, color='red')
sns.lineplot(ax=axes[1], x="Decile", y="cum_event_cap", data= sorted_reindexed, color='blue')
sns.lineplot(ax=axes[1], x="Decile", y="cum_random_cap", data= sorted_reindexed, color='black')
```
